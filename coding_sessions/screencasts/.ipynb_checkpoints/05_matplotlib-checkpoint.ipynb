### Plotting well

- creating a good plot requires a few lines of code
- every plot must have: title and/or legend, axis labels, readable fontsize
- plots should have matching colors
- aspect ratio of plot is also a tool
- personal preference: clear coding structure
- _ax_: **always** 1 axes object
- _axes_: **always** an object holding multiple axes


```python
## creation of data
x = np.arange( 21 )
regression = x * 2
y = x * 2 + np.random.randn( *x.shape)*2
n = 128
image = np.outer( np.arange( n), np.arange( n) ) /n
noise =  np.random.randn( n, n)
noise = 0.6*noise *image.max()/noise.max()
image = image + np.flip( image) + noise
#################################################
#print(x)
#print(x*2)
#print(x.shape)
#print(np.random.randn( *x.shape)*2)
#################################################
#print(np.outer( np.arange( n), np.arange( n) ))
#print(np.outer( np.arange( n), np.arange( n) ) /n)

## Plotting of data
fig, axes = plt.subplots( 1, 2, figsize=(10,4.5))
axes[0].plot( x, regression, color='red', lw=2.5, label='model')
axes[0].scatter( x, y, color='lightblue', edgecolor='k', label='measured data')
axes[1].imshow( image)

## Style layout of plot
titles = [ 'regression', 'noisy color gradients' ]
for ax in axes.flatten():
    ax.set_title( titles.pop( 0 ) )
    ax.grid( ls='--', color='#AAAAAA')
axes[0].legend()
axes[0].set_xlabel( 'x [-]' )
axes[0].set_ylabel( 'f(x) [-]') 
axes[1].axis( 'off' )
```

## Preallocating styles
- colors can be written as tuple
- unpacked dictionaries (_**kwargs_) can be used to allocate linestyles
- axes $\blacktriangleright$ object $\blacktriangleright$ pass into functions

```python
#color = ( R, G, B, alpha)
thick_gray = (0.35, 0.35, 0.35, 0.9) 
vanishing_gray = (0.8, 0.8, 0.8, 0.1)
bright_red = (1, 0, 0, 1)
soothing_blue = (0, 0, 0.8, 0.69) 
strong_blue = ( 0, 0.4, 1)
```

```python
def linestyles( color, lw=2.5, **kwargs):
    """
    Defines multiple lines of the same options with different linestyles
    will return, full line, dashed, dashdot, dotted
    Parameters:
    -----------
    color:      tuple or string
                defined color 
    lw:         float, default 2.5
                linewidth of all lines
    **kwargs:   unpacked tuple
                additional options for the lines
    Returns:
    --------
    lines:      list
                list of lines of the four linestyles
    """
    lines = []
    lines.append( dict( color=color, ls='-', lw=lw, **kwargs) )
    lines.append( dict( color=color, ls='--', lw=lw, **kwargs) )
    lines.append( dict( color=color, ls='-.', lw=lw, **kwargs) )
    lines.append( dict( color=color, ls=':', lw=lw, **kwargs) )
    return lines
```

```python
markers = dict( marker='o', mfc=soothing_blue, mec='k', markersize=6 )
```

```python
def axes_defaults( ax, xlabel, ylabel):
    """
    sets the x and y label of the axes object and adds a grid and legend
    changes the axes object in place
    Parameters:
    -----------
    ax:         plt. axes object
                axes to set the default style
    xlabel:     string
                xlabel of the plot
    ylabel:     string
                ylabel of the plot
    """
    ax.grid( ls=':', color='#AAAAAA' )
    ax.set_xlabel( xlabel)
    ax.set_ylabel( ylabel)
```

```python
## data creation
x = np.arange( 10)
offset = [ 0, -2, 1 ]
scaling = [ 1, 1.5, 0.5 ]
n_lines = len( offset)
print(len( offset))
root_lines = []
square_lines = []
for i in range( n_lines ):
    square_lines.append( x**2 * scaling[i] + offset[i] )
    root_lines.append( 25* x**0.5 * scaling[i] + offset[i] )
#print(square_lines)
#print(root_lines)

fig, ax = plt.subplots(figsize=(8,5))
for i in range( n_lines ):
    ax.plot( x, square_lines[i], lw=2.5 , label='{} x**2 +({})'.format( scaling[i], offset[i]) )
    ax.plot( x, root_lines[i], lw=2.5 , label='{} x**0.5 +({})'.format( scaling[i], offset[i])) 
ax.legend(facecolor=vanishing_gray, edgecolor=strong_blue)
ax.set_xlabel( 'x [-]' )
ax.set_ylabel( 'f(x) [-]') 
```

```python
## linestyle preallocation
red_lines = linestyles( bright_red, **markers)
gray_lines = linestyles( thick_gray )

fig, ax = plt.subplots(figsize=(8,5))
legends = [ 'square function', 'root function'] + 2*(n_lines-1) *[None]
for i in range( n_lines ):
    ax.plot( x, square_lines[i], **gray_lines[i], label=legends.pop(0) )
    ax.plot( x, root_lines[i], **red_lines[i], label=legends.pop(0) ) 
    
plt.rcParams.update( {'font.size':14})
ax.set_title( 'Nonlinear 1D-Functions' )
ax.legend( facecolor=vanishing_gray, edgecolor=strong_blue)
axes_defaults( ax, 'x [-]', 'f(x) [-]' )
```

## File export
- exportable types, vectorgraphic or pixelgraphic
- vectorgraphics WYSWYG 'can zoom infinitely': pdf, eps, svg, etc.<br>
$\quad$ stores 'objects', e.g. straight line $\blacktriangleright$ point coordinate + direction 
- pixelgraphics resolution dependent: png, jpeg, tif, gif <br>
$\quad$ stores pixels, black pixel, white pixel, etc.
- png is lossless
- jpeg is lossy compression via Fourier transform
