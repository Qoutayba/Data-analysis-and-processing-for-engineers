## Accesing dataframe elements

- `pandas.DataFrames` have the methods `loc` and `iloc` for indexing. 
- `df.loc[ index, column_name]` functions with the "index" and the "column names" <br>
$\qquad \to$name location
- names can be seen above with e.g. `df.head()` or `print( df.columns)`
- `df.iloc[ row, column]` uses numbers similar to numpy arrays to access elements <br>
$\qquad \to$ index location
- `df.reindex` is also a method performing the same task

-------
__Task:__ Find the student with the worst grade and print out his attributes using `loc`

```python
grades = df.loc[:,'grade'] 
print('gardes of all Students\n',grades)
#help( grades.idxmax)
worst_student = grades.idxmax()
print(worst_student)
print( '\nThe student with the worst grade' )
print( df.loc[5,'grade'] )
```

## Treating missing numbers (nan)
- `nan` usually has to be replaced to further use the data
- popular ways to fill out `nan` are: median, mean or 0<br>
- pandas offers multiple methods on filling `nan` <br>
$\qquad \to$ specify columns, fill with specified value

-------
__Task:__ Fill out the remaining `nan` in grade and semester with the median values of the columns

```python
df = pd.read_csv( 'data/student_info.csv' )
df = df.fillna( {'attendance %':df['attendance %'].mean(), 'hours studied':df['hours studied'].mean() })
df = df.fillna( {'grade':df['grade'].median(), 'semester':df['semester'].median() })


check.nan( df)
```

## Manipulating values in dataframes


Since corona the students have much more time to attend the lecture, hence the column 'attendance %' does not reflect the reality anymore.

-------
__Task:__ Increase the 'attendance %' of all the students by 10%. Make sure it does not go above 100%

```python
df = pd.read_csv( filename) 
#print( df.head(37) )
df.loc[:, 'attendance %'] + (df.loc[:, 'attendance %']*0.1)
df.loc[:, 'attendance %'] = np.minimum( df.loc[:, 'attendance %'] + (df.loc[:, 'attendance %']*0.1), 100 )
#print(attendance)
#print( df.head(37) )
check.attendance( df)
```

- columns can be added to dataframes similar to python dictionaries <br>
- recall `dict[new_key] = values_ expands the _dict`<br>
$\qquad\to$ analogously  `df['column'] = data`  works
- `data` should be a dictionary, i.e. `{ column_name : values}`
- Note that `df.append( dict/df)` adds rows and not columns
