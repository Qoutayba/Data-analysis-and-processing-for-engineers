## Plotting guidelines
- plots are very easy to mess up
- each plot must have the following:<br>
$\quad$ title, axis labels with units, matching aspect ratio, correct plotstyle (scatter, line, line with markers, etc.)
- each plot should have (not always a must)<br>
$\quad$ grid, legend, zero on y-axis
- a plot should be self contained, additional description can be given in e.g. in the caption or on the text
- sometimes less is more, too much information/too descriptive text makes the plot cluttered and confusing
- a plot should give a coarse overview/understanding of _what_ is presented on first glance

## Creating a plot with zoomout

- a region of interest can be highlighted with an inset plot
- the section can be framed with e.g. putting another `axes` inside the `axes` 
- the inner `axes` should have no ticks, bigger spines, translucent facecolor
- `mark_inset` can be helpful
- a few examples are given in the 'example_plots.py' script
- for a top notch plot, usually additional data processing is required
- __Note:__ To compute the intersection additional processing on e.g. paper or finer sampling of $x$ is required

-----------
__Task:__ Enhance the plot, highlight the intersections in another window. <br>
Compute the intersection points and mark them with scatters.

```python
from matplotlib import rc
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

x = np.arange( 0, 40, 0.2)
square_line = lambda x, a=0.1, c=5.1: a*x**2 + c
log_line    = lambda x, a=5.0, c=0.5: a*np.log( np.maximum(0.01, x) ) + c
sine_line   = lambda x, a=1.9, b=1.5, c=-4.8: a*x + b* np.sin( x) + c

line_data = []
line_data.append( square_line(x) )
line_data.append( square_line(x, a=0.15, c=7) )
line_data.append( log_line(x) )
line_data.append( log_line(x, a=4.0, c=-4) )
line_data.append( sine_line(x) )

line_labels = []
line_labels.append( 'first square_line(x)' )
line_labels.append( 'second square_line(x)' )
line_labels.append( 'first log_line(x)' )
line_labels.append( 'second log_line(x)' )
line_labels.append( 'sine_line(x)' )


fig, axes = plt.subplots(figsize=(12,7))
for line,labels in zip(line_data,line_labels):
    axes.plot( x, line, label=labels , lw=2.5)

    
axes.set_xlim( x.min(), x.max() )
axes.set_title( 'plot with zoomout' )
axes.legend(fontsize='large')
axes.set_xlabel( 'x [-]',fontsize='x-large' )
axes.set_ylabel( 'f(x) [-]',fontsize='x-large' )
axes.grid( ls=':', color='#AAAAAA', linewidth=2.5 )


## adding of inset
box_color = (0.85, 0.85, 0.85, 0.3)
box = plt.axes( (0.25, 0.46, 0.25, 0.2) ) 
box.plot( x, line_data[0], color='orange', lw=2 )
box.plot( x, line_data[1], color='red', lw=2 ) 
box.plot( x,line_data[2], color='blue', lw=2, ls='--' )
box.plot( x,line_data[3], color='green', lw=2, ls='-' )
box.plot( x,line_data[4], color='k', lw=2, ls=':' )

mark_inset( axes, box, loc1=2, loc2=4, facecolor=box_color, edgecolor='k', lw=1.5 )

box.set_xlim( xmin=0.2, xmax=9)
box.set_ylim( ymin=-20, ymax=20 ) 
box.text( 4, -5, 'intersection', horizontalalignment='center', verticalalignment='center')
box.tick_params( axis='y', which='both', left=False, labelleft=False)
plt.setp( box, xticks=[], yticks=[] )

plt.show()
```
