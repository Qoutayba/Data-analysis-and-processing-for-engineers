```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Example 1 - labeling, line style, data ranges
- label *all* axes including physical units (use [-] if none/de-dimensionalized quantities are plotted)
- using sufficiently large font size
- disambiguate multiple lines in a sensible way
- if possible, choose colors that are distinguishable on b/w plots
- make the plot easy to read $\blacktriangleright$ more attention for the content

```python
x = np.linspace(-1,1,1000)
y1= 1.0 - x*x
y2= np.cos(np.pi * x/2)

fig = plt.figure(figsize=(5,5))
ax = plt.gca()
ax.set_title('Not ideal')
ax.plot( x, y1, color='black')
ax.plot( x, y2, color='black')

fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].set_title(r'Much better: $x^2$ vs. $\cos(\pi x/2)$', fontsize=16)
ax[0].plot( x, y1, color='black', linewidth=3)
ax[0].plot( x, y2, color='dodgerblue', linewidth=3, dashes=[6, 4])
ax[0].set_xlabel(r'$x$ [-]',fontsize=14)
ax[0].set_ylabel(r'$f(x), g(x)$ [-]',fontsize=14)
ax[0].grid()
ax[0].legend([ r'$f(x)=x^2$', r'$g(x)=\cos(\pi x/2)$'], fontsize=14)
ax[1].set_title(r'Difference $g(x)-f(x)$', fontsize=16)
ax[1].plot( x, y2-y1, color='red', linewidth=2)
ax[1].set_xlabel(r'$x$ [-]',fontsize=14)
ax[1].set_ylabel(r'$g(x)-f(x)$ [-]',fontsize=14)
ax[1].grid()
plt.tight_layout()
```

## Example 2: discrete data should remain discrete

- use e. g. scatter or bar plots when showing discrete data
- set bar width in a sensible way
- **hint:** modify opacity using 'alpha'

### Example 2a: The 6-sided dice
- probability for numbers $k \in \{ 1, \dots, 6\}$ are all $P_k=1/6$
- continuous plotting could suggest intermediate values to exist and have same probability
- probability is nonnegative and $\leq 1$; plot range should account for that

```python
P = np.ones(6)/6
x = np.arange(1,7)
fig, ax= plt.subplots(1,2,figsize=(10,5))

ax[0].bar( x, P, width=0.4, color='blue' )
ax[0].set_ylim( (0, 0.2))
ax[0].set_xlabel(r'$k$ [-]', fontsize='large')
ax[0].set_ylabel(r'$P_k$ [-]', fontsize='large')
ax[0].set_title('6-sided dice: discrete probabilities', fontsize='x-large')
ax[1].plot(x,P, color='blue')

plt.tight_layout(pad=5.0)
```

## Example 3: Bad plot range

- goes in the direction of journalistic vs. scientific plotting
- having the wrong field of view can bias the one way or the other

```python
n_bin = 10
n_sample = 30000
u = np.linspace(0,1, n_bin+1)
l = u[:-1]
u = u[1:]
w = u[0]-l[0]
c = 0.5*(u+l)
y = np.sort( np.random.uniform( 0, 1, size=n_sample) )
N = np.zeros( n_bin, dtype='long')
i_bin = 0

# naive implementation of the binning using sorted data:
for i_sample in range(n_sample):
    while y[i_sample] > u[i_bin]:
        i_bin += 1
    N[i_bin] += 1

rel_freq = N/n_sample
cum_freq = rel_freq.cumsum()
fig, ax = plt.subplots(1,2,figsize=(10,5))

# plot the data with a correct view on the data
AX = ax[0]
AX.set_title('rel. frequency', fontsize='x-large')
AX.set_xlabel(r'$x$ [-]', fontsize='large')
AX.set_ylabel(r'$P_k$ [-]', fontsize='large')
AX.bar( c,rel_freq, width=0.75*w, color='cornflowerblue', alpha=0.6, label = 'rel. frequency', zorder=10 )
AX.plot( (0,1), (1/n_bin, 1/n_bin), color='black', linewidth=2, label='theoretical value', zorder=5 )
AX.set_ylim((0, np.max(rel_freq) + 0.05/n_bin) )

AX.legend(loc='lower center', shadow=True,facecolor='white',fontsize='large', framealpha=1).set_zorder(15)

# plot the data focusing only on deviations
# -> can lead to biased interpretations
AX = ax[1]
AX.set_title('biased view on the data', fontsize='x-large')
AX.set_xlabel(r'$x$ [-]', fontsize='large')
AX.set_ylabel(r'$P_k$ [-]', fontsize='large')
AX.bar( c,rel_freq, width=0.75*w, color='cornflowerblue', alpha=0.6, label = 'rel. frequency', zorder=10 )
AX.plot( (0,1), (1/n_bin, 1/n_bin), color='black', linewidth=2, label='theoretical value', zorder=5 )
AX.set_ylim((np.min(rel_freq)-1/n_sample, np.max(rel_freq)+1/n_sample ) )

AX.legend(loc='best', shadow=True,facecolor='white',fontsize='large', framealpha=1).set_zorder(15)

plt.tight_layout(pad=1.1)
```
